```python
# Importando bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importando dados
menu = pd.read_csv('dados/menu.csv')
```

```python
menu.head()
```

```python
menu.tail()
```

```python
# Verificando valores nulos
menu.isnull().sum()
```

```python
# Filtrando valores médios de interesse por categoria
categorias = menu.groupby('Category',as_index=False)[['Calories','Total Fat','Sugars','Sodium','Carbohydrates','Protein']].mean()
```

```python
# Ordenando por calorias em ordem crescente
categorias.sort_values(by='Calories',inplace=True)
```

```python
categorias
```

##### Média de  calorias por categoria

```python
# Utilizando paleta de cores do seaborn
cores = sns.color_palette(palette='plasma')
# Criando Gráfico horizontal
plt.barh(categorias['Category'],categorias['Calories'],color=cores)
plt.title('Média de calorias por categoria de alimento')
plt.xlabel('Calorias')
plt.ylabel('Categoria')
plt.show()
```

##### Média de  sódio por categoria

```python
# Criando Gráfico horizontal
plt.barh(categorias['Category'],categorias['Sodium'],color=cores)
plt.title('Média de  sódio por categoria de alimento')
plt.xlabel('Sódio')
plt.ylabel('Categoria')
```

##### Informações nutricionais médias por categoria

```python
plt.rcParams.update({'font.size': 7})
fig, ax = plt.subplots(figsize=(11, 6))

bar_width = 0.15
#Determinando posição das barras
bar_positions = range(len(categorias['Category']))
for i, nutrient in enumerate(categorias.columns[1:]):
    ax.bar([pos + (bar_width * i) for pos in bar_positions], categorias[nutrient], label=nutrient, width=bar_width,color=cores[i % len(cores)])


ax.set_xticks([pos + (bar_width * (len(categorias.columns) - 1) / 2) for pos in bar_positions])
ax.set_xticklabels(categorias['Category'])
ax.set_xlabel('Categoria de Alimento')
ax.set_ylabel('Quantidade Média')
ax.set_title('Informações Nutricionais médias por Categoria de Alimento')
ax.legend()
#Adicionando Grid
ax.grid(axis = 'y')
#Salvando gráfico
plt.savefig('img/info_nutri.jpg')

# Exibindo o gráfico
plt.show()


```

##### Criando um cardápio diário baseado nas categorias

```python
manha = categorias.loc[categorias['Category'].isin(['Breakfast', 'Coffee & Tea'])].sum()
almoco = categorias.loc[categorias['Category'].isin(['Beef & Pork', 'Beverages','Salads','Desserts'])].sum()
tarde = categorias.loc[categorias['Category'].isin(['Snacks & Sides', 'Smoothies & Shakes'])].sum()
noite = categorias.loc[categorias['Category'].isin(['Chicken & Fish', 'Beverages','Salads'])].sum()

total = manha + almoco + tarde + noite
```

```python
# Criando dataframe com os valores do total
mcdonalds = pd.DataFrame({
    'Nutrientes': ['Calories','Total Fat','Sugars','Sodium','Carbohydrates','Protein'],
    'Valores': [3624.272433,143.246948, 241.488772,5582.523756,438.898151,150.739837]
     
})
```

```python
# Criando dataframe com nutrientes diários recomendados (Homem adulto) fonte -> https://www.verywellfit.com
recomendado = pd.DataFrame({
    'Nutrientes': ['Calories','Total Fat','Sugars','Sodium','Carbohydrates','Protein'],
    'Valores': [2000.00,73.00,50.00,2300.00,255.00,90.00]
     
})
```

```python
#Realizando Inner Join
comparacao = mcdonalds.merge(recomendado,on='Nutrientes',suffixes=('_Mc','_Reco'))
```

```python
comparacao
```

```python
# Configurando o gráfico
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = range(len(comparacao))

# Gráfico de barras para 'Valores_Mc'
bar1 = ax.bar(index, comparacao['Valores_Mc'], bar_width, label='Valores_Mc',color='#8807A5')

# Gráfico de barras para 'Valores_Reco'
bar2 = ax.bar([i + bar_width for i in index], comparacao['Valores_Reco'], bar_width, label='Valores_Reco',color='#F9D924')

# Configuração do eixo x
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(comparacao['Nutrientes'])
ax.set_xlabel('Nutrientes')
ax.set_ylabel('Valores')
# Adicionando valores
plt.bar_label(bar1,fontsize=7,label_type='edge',fontweight='bold')
plt.bar_label(bar2,fontsize=7,label_type='edge',fontweight='bold')
# Adicionando a legenda
ax.legend()
#Adicionando Grid
ax.grid(axis = 'y')
#Salvando Gráfico
plt.savefig('img/recoxmc.jpg')

# Exibindo o gráfico
plt.show()
```
